```python
plt.figure(figsize=(5,3))
plt.bar(x,y, color='r')
plt.xlabel('representing my cat data')
plt.ylabel('representing my num data')
plt.title("this is my bar plot")
plt.grid()## with the help of bar we can graph categorial vs numerical
```
