# merge and join operations

```python
data1 = pd.DataFrame( { "key1" : [2,3,4,5],
      "key2" : [4,5,6,7],
      "key3": [2,3,4,5]})
```

```python
data2 = pd.DataFrame( { "key1" : [2,3,34,5],
      "key5" : [4,57,6,7],
      "key6": [2,43,4,5]})
```

```python
data1 
```

```python
data2
```

```python
# Merging 
pd.merge(data1 , data2)
```

```python
pd.merge(data1 , data2 , how = 'left')
```

```python
data1
```

```python
pd.merge(data1 , data2 , how = 'right')
```

```python
pd.merge(data1 , data2 , how = 'outer')
```

```python
pd.merge(data1 , data2 , how = 'cross')
```

```python
pd.merge(data1 , data2 , how = 'left' , on = 'key1')
```

```python
pd.merge(data1 , data2 , how = 'left' , left_on = 'key2' , right_on = 'key5')
```
